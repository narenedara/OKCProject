```python

import csv
import numpy as np
import pandas as pd
# import numpy
df=pd.read_csv("shots_data.csv")
df["Hypotenuse"]=np.sqrt((df["x"]**2)+(df["y"]**2))
two_point_a_total=0
two_point_a_make=0
two_point_b_total=0
two_point_b_make=0
non_corner_three_totala=0
non_corner_three_makea=0
non_corner_three_totalb=0
non_corner_three_makeb=0
corner_three_totala=0
corner_three_makea=0
corner_three_totalb=0
corner_three_makeb=0
a_total_shots=0
b_total_shots=0
count=0
for ind in df.index:
    if df["team"][ind]=="Team A":
        a_total_shots+=1
        if (df['y'][ind]<=7.8 and -22.0 <df['x'][ind]<22.0 ) or (df['y'][ind]>7.8 and df['Hypotenuse'][ind]<23.75):
             two_point_a_total+=1
             if (df['fgmade'][ind]==1):
                two_point_a_make+=1
        if (df['y'][ind]>7.8 and df['Hypotenuse'][ind]>23.75):
             non_corner_three_totala+=1
             if (df['fgmade'][ind]==1):
                non_corner_three_makea+=1
        if (df['y'][ind]<=7.8 and (df['Hypotenuse'][ind]>22.0 or df['Hypotenuse'][ind]<-22.0)):
             corner_three_totala+=1
             if (df['fgmade'][ind]==1):
                corner_three_makea+=1
    if df["team"][ind]=="Team B":
        b_total_shots+=1
        if (df['y'][ind]<=7.8 and -22.0 <df['x'][ind]<22.0 ) or (df['y'][ind]>7.8 and df['Hypotenuse'][ind]<23.75):
             two_point_b_total+=1
             if (df['fgmade'][ind]==1):
                two_point_b_make+=1
        if (df['y'][ind]>7.8 and df['Hypotenuse'][ind]>23.75):
             non_corner_three_totalb+=1
             if (df['fgmade'][ind]==1):
                non_corner_three_makeb+=1
        if (df['y'][ind]<=7.8 and (df['Hypotenuse'][ind]>22.0 or df['Hypotenuse'][ind]<-22.0)):
             corner_three_totalb+=1
             if (df['fgmade'][ind]==1):
                corner_three_makeb+=1
            

percent_two_point_totala= two_point_a_total / a_total_shots
percent_two_point_totalb= two_point_b_total / b_total_shots
percent_non_corner_threea= non_corner_three_totala / a_total_shots
percent_non_corner_threeb= non_corner_three_totalb / b_total_shots
percent_corner_threea=corner_three_totala / a_total_shots
percent_corner_threeb=corner_three_totalb / b_total_shots



efg_twopt_a= (two_point_a_make/two_point_a_total)
efg_noncornerthree_a= (non_corner_three_makea+(0.5*non_corner_three_makea))/non_corner_three_totala
efg_cornerthree_a= (corner_three_makea+(0.5*corner_three_makea))/corner_three_totala
efg_twopt_b= (two_point_b_make/two_point_b_total)
efg_noncornerthree_b= (non_corner_three_makeb+(0.5*non_corner_three_makeb))/non_corner_three_totalb
efg_cornerthree_b= (corner_three_makeb+(0.5*corner_three_makeb))/corner_three_totalb

print("% of total shots that are 2pts(A):"+str(percent_two_point_totala))
print("% of total shots that are 2pts(B):"+str(percent_two_point_totalb))
print("% of total shots that are Non-Corner 3pts(A):"+str(percent_non_corner_threea))
print("% of total shots that are Non-Corner 3pts(B):"+str(percent_non_corner_threeb))
print("% of total shots that are Corner 3pts(A):"+str(percent_corner_threea))
print("% of total shots that are Corner 3pts(B):"+str(percent_corner_threeb))
print("Efficient FG percentage 2pt(A):"+str(efg_twopt_a))
print("Efficient FG percentage Non-Corner 3pt(A):"+str(efg_noncornerthree_a))
print("Efficient FG percentage Corner 3pt(A):" +str(efg_cornerthree_a))
print("Efficient FG percentage 2pt(B):"+str(efg_twopt_b))
print("Efficient FG percentage Non-Corner 3pt(B):"+str(efg_noncornerthree_b))
print("Efficient FG percentage Corner 3pt(B):"+ str(efg_cornerthree_b))






    
    
        
```
